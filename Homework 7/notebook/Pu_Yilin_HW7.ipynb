```python
np.logspace(start = -2,stop = 3, num = 6)

tuned_param = [{ 'gamma': np.arange(.2,2.1,.1), 'C': np.logspace(start = -2,stop = 3, num = 6)}]

def chooseParam( train_x, train_y, test_x, test_y, model = SVC(kernel='rbf'),tuned_params=tuned_param):
    model = GridSearchCV(estimator=model, param_grid = tuned_param, scoring = 'f1_weighted', n_jobs = -1,\
                       cv = StratifiedKFold(10, random_state = 100, shuffle=True) )
    
    model.fit(train_x, train_y)
    
    print("Grid scores are: ")
    
    
    means = model.cv_results_['mean_test_score']
    stds = model.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, model.cv_results_['params']):
        print("%0.2f (+/-%0.02f) for %r"
              % (mean, std * 2, params))
    print("The best parameters are: ",model.best_params_)
    print("The best score for the SVM is: ",model.best_score_)

    
    test_pred = model.predict(test_x)
    print("The classification report: ",classification_report(test_y, test_pred))
    return model
```

```python
gaussian_family = chooseParam(trainx, trainy['Family'], testx, testy['Family'])

gaussian_genus = chooseParam(trainx, trainy['Genus'], testx, testy['Genus'])

gaussian_species = chooseParam(trainx, trainy['Species'], testx, testy['Species'])
```

```python
def AllDisplay(model, train_x, train_y, test_x, test_y, label ):

    ova_svm = OneVsRestClassifier(SVC(kernel='rbf', C = model.best_params_['C'], gamma = model.best_params_['gamma'], probability=True))
    ova_svm_fit =  ova_svm.fit(train_x, train_y)
    

    pred_y_train = ova_svm_fit.predict(train_x)
    print("The confusion matrix of the training data ",label," is \n",confusion_matrix(train_y, pred_y_train))
    print("F1-score of the training data %s is \n%0.3f "  %(label, f1_score(train_y, pred_y_train, average='weighted'))) 
    

    pred_y_test = ova_svm_fit.predict(test_x)
    print("The confusion matrix of the testing data ",label," is \n",confusion_matrix(test_y, pred_y_test))
    print("F1-score of the testing data %s is \n%0.3f "  %(label , f1_score(test_y, pred_y_test, average='weighted'))) 
    

    print("The hamming loss of the testing data ",label," is \n",hamming_loss(test_y, pred_y_test))
    print("The exact match of the testing data ",label," is \n",\
          accuracy_score(test_y, pred_y_test))
    return test_y, pred_y_test
```

```python
test_y_fam, pred_y_fam = AllDisplay(gaussian_family, trainx, trainy['Family'], testx, testy['Family'],'Family')

test_y_gen, pred_y_gen = AllDisplay(gaussian_genus, trainx, trainy['Genus'], testx, testy['Genus'],'Genus')

test_y_spec, pred_y_spec = AllDisplay(gaussian_species, trainx, trainy['Species'], testx, testy['Species'],'Species')
```

### 2. K-Means Clustering on a Multi-Class and Multi-Label Data Set
#### Monte-Carlo Simulation: Perform the following procedures 50 times, and report the average and standard deviation of the 50 Hamming Distances that you calculate.

#### (a) Use k-means clustering on the whole Anuran Calls (MFCCs) Data Set (do not split the data into train and test, as we are not performing supervised learning in this exercise). Choose kâˆˆ {1,2, . . . ,50} automatically based on one of the methods provided in the slides (CH or Gap Statistics or scree plots or Silhouettes) or any other method you know.

```python
def Silhoutte(kmax, Xtrain, rand= 50 ):
    best_K = 2
    best_score = 0
    sil_score = np.array([])
    k_val = np.array([])
    for k in range(2, kmax+1):
        k_val = np.append(k_val, k)
        kmeans = KMeans(n_clusters = k, random_state = rand)
        labels = kmeans.fit_predict(Xtrain)
        cur_score = silhouette_score(Xtrain, labels, metric = 'euclidean')
        print("For ", k," clusters, the average silhouette score is: ", cur_score)
        sil_score  = np.append(sil_score, cur_score)
    ind = np.argmax(sil_score)
    best_K = k_val[ind]
    
    print("The best K is: {}".format(best_K))
    return best_K, sil_score
```

```python
best_k, sil_score = Silhoutte(50, data_x)
```

#### (b)In each cluster, determine which family is the majority by reading the true labels. Repeat for genus and species.

```python
def themajority(label, all_labels):
    clusters = list(set(all_labels))
    clusters.sort()
    majority = np.array([])

    for cur_clus in clusters:
        
        index, = np.where(all_labels == cur_clus)
        cluster_item = label[index]
        majority = np.append(majority, cluster_item.value_counts().idxmax())
    majority = pd.Series(majority)
    return majority 
```

```python
km = KMeans(n_clusters = best_k.astype(int), random_state = 0)
cluster_lab = km.fit_predict(data_x)

majority_fam = themajority( data_y['Family'], cluster_lab)
print("Majority in Family label is \n",majority_fam)

majority_gen = themajority( data_y['Genus'], cluster_lab)
print("Majority in Genus label is \n",majority_gen)

majority_spe = themajority( data_y['Species'], cluster_lab)
print("Majority in Species label is \n",majority_spe)
```

#### (c) Now for each cluster you have a majority label triplet (family, genus, species). Calculate the average Hamming distance, Hamming score, and Hamming loss between the true labels and the labels assigned by clusters.

```python
def getPred( majority, all_labels):
    clusters = list(set(all_labels))
    
    pred_lab = np.empty(len(all_labels), dtype="object")

    for cur_clus in clusters:
        index, = np.where(all_labels == cur_clus)
        pred_lab[index] = majority[cur_clus]
    return pred_lab 
```

```python
label_fam = getPred( majority_fam, cluster_lab)
print(label_fam)
label_gen = getPred(majority_gen, cluster_lab)
print(label_gen)
label_spe = getPred(majority_spe, cluster_lab)
print(label_spe)
```

```python

def hamm_loss(true_arr, pred_arr, avg_dist = False):
    sums = np.sum(np.not_equal(true_arr[2],pred_arr[2])) + \
    np.sum(np.not_equal(true_arr[1],pred_arr[1])) + np.sum(np.not_equal(true_arr[0],pred_arr[0]))
    denom = true_arr.shape[0] * true_arr.shape[1]
    ham_loss = sums/denom
    
    if avg_dist == True:
        avg_ham_dist = sums/true_arr.shape[1]
        return avg_ham_dist, ham_loss
    
    return ham_loss
```

```python
true_arr_k = np.vstack([data_y['Family'],data_y['Genus'],data_y['Species']])
pred_arr_k =  np.vstack([label_fam, label_gen, label_spe] )

avg_dist,hamming_loss = hamm_loss(true_arr_k, pred_arr_k, True)
print("The avg hamming distance: ",avg_dist," and hamming loss is ",hamming_loss)

print("The avg Hamming score is",(1 - np.mean(hamming_loss)))
```

```python
def monteCarlo(n, Xtrain, ytrain  ):
    true_arr_mc = np.vstack([ytrain['Family'], ytrain['Genus'], ytrain['Species']])
    ham_all = np.array([])
    ham_dist = np.array([])
    for i in range(n):
        
        best_k, sil_score = Silhoutte(n, Xtrain, i) 

        km = KMeans(n_clusters= best_k.astype(int), random_state = 50)
        cluster_lab = km.fit_predict(Xtrain)

        majority_fam = themajority( ytrain['Family'], cluster_lab)

        majority_gen = themajority( ytrain['Genus'], cluster_lab)

        majority_spe = themajority( ytrain['Species'], cluster_lab)
        
        label_fam = getPred( majority_fam, cluster_lab)
        label_gen = getPred(majority_gen, cluster_lab)
        label_spe = getPred(majority_spe, cluster_lab)
        pred_arr_k =  np.vstack([label_fam, label_gen, label_spe] )

        avg_ham_dist, ham_lossmc = hamm_loss(true_arr_k, pred_arr_k, True)
        print("For iteration[", i, "] The hamming loss is:", best_k, " is", ham_lossmc)
        print("For iteration[", i, "] The hamming dist is:", best_k, " is", avg_ham_dist)

        ham_all = np.append(ham_all, ham_lossmc) 
        ham_dist = np.append(ham_dist, avg_ham_dist) 
    return ham_dist, ham_all
```

```python
all_dist, all_hamming = monteCarlo(50 ,data_x, data_y)
```

```python
print("Average Hamming Loss: ", np.mean(all_hamming))
print("Std Hamming Loss:", np.std(all_hamming))

print("Average Hamming distance: ", np.mean(all_dist))
print("Std Hamming distance: ", np.std(all_dist))

print("Average of Hamming score: ",(1 - np.mean(all_hamming)))
print("Std Hamming score:  ",np.std(all_hamming))
```
